```python
import camelot
```

```python
def load_data(file_path,pages):
    all_tables=[]
    for page in pages:
        tables = camelot.read_pdf(file_path,pages=str(page),flavor="stream")
        for table in tables:
            table_df = table.df
            table_df = ( table_df.rename(columns=table_df.iloc[2]))
            
            df = table_df[3:]
            all_tables.append(df)
    return all_tables
            
```

```python
file_path ="./data/2015-cocacola-10k-filing.pdf"
pages =range(75,77)

tables =load_data(file_path,pages)
```
